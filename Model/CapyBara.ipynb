```python
columns_to_scale = ['LATITUDE', 'LONGITUDE', 'pH', 'EC', 'CO3', 'HCO3', 'Cl', 'SO4', 'NO3',
       'PO4', 'TH', 'Ca', 'Mg', 'Na', 'K', 'F', 'U(ppb)']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
X=df.drop(['pH', 'EC', 'CO3', 'HCO3', 'Cl', 'SO4', 'NO3', 'PO4', 'TH', 'Ca', 'Mg', 'Na', 'K', 'F', 'U(ppb)'],
         axis=1)
```

```python
y=df[['pH', 'EC', 'CO3', 'HCO3', 'Cl', 'SO4', 'NO3', 'PO4', 'TH', 'Ca', 'Mg', 'Na', 'K', 'F', 'U(ppb)']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
X_train
```

```python
y_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X.shape
```

```python
y.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=42, activation='relu', input_shape=(42,)),
    tf.keras.layers.Dense(21, activation='relu'),
    tf.keras.layers.Dense(15, activation='softmax')
])

model.compile(optimizer='adam', 
              loss=tf.losses.mean_squared_error, 
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
predictions = model.predict(X_test)
```
